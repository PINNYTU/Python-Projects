# Import library 

```python
import pandas as pd
import datetime 
import time
import numpy as np
import matplotlib as mp
```

# Import Data .csv 

```python
raw_df=pd.read_csv("Absenteeism-data.csv")
raw_df
```

```python
list(raw_df)
```

# Start Coding drop columns

```python
raw_df.drop(columns="ID",inplace= True)
df_group=raw_df
df_group
```

# Create new variables for each values in a column

```python
reason_group=pd.get_dummies(df_group["Reason for Absence"])
reason_group.columns.values
```

```python
reason_group
reason1=reason_group.loc[:,1:14].max(axis=1)
reason2=reason_group.loc[:,15:17].max(axis=1)
reason3=reason_group.loc[:,18:21].max(axis=1)
reason4=reason_group.loc[:,22:28].max(axis=1)
new_df=pd.concat([df_group,reason1,reason2,reason3,reason4],axis=1)
new_df
```

```python
new_df.columns.values
```

```python
column_name= ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3','Reason4']
new_df.columns=column_name
C_new_df=new_df.copy()
new_df
```

```python
new_df.drop(columns="Reason for Absence")
```

# Adjust datetime (Get only the needed time)

```python
testdate=new_df["Date"]
print(testdate)
```

```python
day=[]
month=[]
dayformat="%d"
monthformat="%m"
for d in testdate :
    d = datetime.datetime.strptime(d,"%d/%m/%Y")
    day.append(datetime.date.strftime(d, dayformat ))
    month.append(datetime.date.strftime(d, monthformat))  
day = pd.DataFrame(day)
month = pd.DataFrame(month)
print(day)
print(month)
```

```python
new_df=pd.concat([new_df,day,month],axis=1)
```

```python
new_df.columns.values
```

```python
new_df.columns= ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3',
       'Reason4', 'Day of week', 'Month value']
new_df.drop(columns='Date',inplace=True)
```

```python
D_new_df=new_df.copy()
new_df
```

# Change values in column

```python
x=[]
for d in new_df["Education"] :
    if d==1 :
        x.append(0)
    if d>1 :
        x.append(1) 
Education=pd.DataFrame(x,columns=['Education'])
Education
```

```python
new_df.drop(columns='Education',inplace=True)
new_df=pd.concat([new_df,Education],axis=1)
new_df
```
